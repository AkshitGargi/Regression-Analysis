```python
print(df[df['Pclass'] == 1]['Age'].mean())
print(df[df['Pclass'] == 2]['Age'].mean())
print(df[df['Pclass'] == 3]['Age'].mean())
```

```python
def age_impute(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 38.23
        elif Pclass == 2:
            return 29.87
        else:
            return 25.14
    
    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass' ]].apply(age_impute, axis = 1)
```

```python
df = df.drop(['PassengerId'], axis = 1)
```
