```python
#Writing a function to impute the age

def age_impute(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 38.23
        elif Pclass == 2:
            return 29.87
        else:
            return 25.14
    
    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass' ]].apply(age_impute, axis = 1)
```

```python
df = df.drop(['Name', 'Ticket'], axis = 1)
```

```python
df['Sex'] = pd.get_dummies(df.Sex)['female']
```

```python
df = pd.get_dummies(df, columns=["Embarked"])
df
```

```python
df = df.drop(['Embarked_S'], axis = 1)
df
```

```python
y = df.pop('Survived')
y
```

```python
X = df
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y ,random_state = 123)
```

```python
from sklearn.linear_model import LogisticRegression


model = LogisticRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
matrix = metrics.confusion_matrix(y_test,model.predict(X_test))
print(matrix)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,model.predict(X_test)))
```
